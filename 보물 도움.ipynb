```python
import random

# 게임 초기화
def initialize_game(n):  # 랜덤으로 (0 ~~ 4) 범위중에 숫자 부여해줌
    board_size = n # 5
    treasure_position = (random.randint(0, n-1), random.randint(0, n-1)) # 보물의 위치 (x,y)x는 행 y 열
    player_position = (random.randint(0, n-1), random.randint(0, n-1)) # 플레어의 위치 (x,y)
    return board_size, treasure_position, player_position # 튜플 행 과 열로 표현하는데 쓸 수 있음 

# 플레이어 이동
def move_player(board_size, player_position, direction): #2차원 배열마냥 위치를 조정함
    x, y = player_position 
    if direction == 'N': # 북 x행 감소시 위쪽으로 이동
        x = max(x-1, 0)  # max 보드 경계를 벗어나지 않도록 최소 0이 되도록함,  ex)(3,0) 감소하면 (2,0) 3보다 위에 있으니 위로 이동
    elif direction == 'S': # 남 x행 증가시 아래쪽으로 이동
        x = min(x+1, board_size-1) # min을 쓰는 이유 x좌표를 증가 시키되 최대값이 보드 크기 -1로 제한 하기 위해서
    elif direction == 'E': # 동 y행 증가시 오른쪽 이동
        y = min(y+1, board_size-1) # min board_size-1이 되게함  ex) (4,4) y+1을 하면 (5,4)가 되지만 min함수로 4만 남고 y에는 4가 할당 보드 크기를 넘길수 없음
    elif direction == 'W':# 서 y행 감소시 왼쪽으로 이동
        y = max(y-1, 0) # max를 이용 y좌표를 감소 시키되 최소값은 0 이 되도록함  
    return x, y  # 새로운 위치 반환

# 거리 계산
def calculate_distance(treasure_position, player_position): # abs() 함수 절대값구해줌 
    return abs(treasure_position[0] - player_position[0]) + abs(treasure_position[1] - player_position[1]) # 함수를 호출한 distance로 값을 보내줌
    # x의 보물위치 플레어 위치 y의 보물 플레어 위치를 빼고 더해주면 거리가 나옴 , 맨해튼 거리 계산 공식 가로 세로로만 이동하는 경우 사용하며 단순 보드게임 경로찾기 문제에서 자주 사용됨

# 게임 실행
def play_game(board_size): # '_' (언더스코어) 값을 무시하고 싶은 경우 board_size의 값은 이미 알고 있음 initialize_game에서 할당값을 굳이 받을 필요가 없으므로 사용
    _, treasure_position, player_position = initialize_game(board_size) # 보드 사이즈 5를 n으로 보냄
    moves = 0 # 이동 횟수 추척하는데 사용됨
    
    while True:
        print(f"현재 위치: {player_position}")
        distance = calculate_distance(treasure_position, player_position)
        print(f"보물까지의 거리: {distance}")
        
        if distance == 0:
            print(f"보물을 찾았습니다! 이동 횟수: {moves}")
            break # while문을 빠져나옴

        try:
            direction = input("이동 방향을 선택하세요 (N, S, E, W): ").upper() # 입력값 대문자로 변환해줌
            if direction not in ['N', 'S', 'E', 'W']: # N,S,E,W 가 아닌경우 ValueError 발생시킴
                raise ValueError("잘못된 방향입니다. N, S, E, W 중 하나를 입력해주세요.")
            player_position = move_player(board_size, player_position, direction) # 변경 위치를 플레어 위치에 할당
            moves += 1
        except ValueError as e:
            print(e) # '잘못된 방향입니다 ~ 중 하나를 입력해주세요' 출력 해줌

# 게임 보드 크기 설정 및 게임 시작
if __name__ == "__main__":
    board_size = 5  # 보드 크기를 5x5로 설정
    play_game(board_size)
```
