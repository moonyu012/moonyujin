[과제 2: 초기화 메서드 __init__]

목표: __init__ 메서드를 구현하여 붕어빵 가게의 초기 재고, 가격, 판매량을 설정합니다.

설명: self.stock, self.prices, self.sales 딕셔너리를 초기화하세요.
- stock 
    - "팥붕어빵": 10
    - "슈크림붕어빵": 8
    - "초코붕어빵": 5

- prices
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- sales
    - "팥붕어빵": 0,
    - "슈크림붕어빵": 0,
    - "초코붕어빵": 0

```python
class BungeoppangShop:
    def __init__(self):
        self.check_stock = {
            "팥붕어빵": 10,
            "초코붕어빵": 8,
            "슈크림붕어빵": 5
        }
        self.prices = {
            "팥붕어빵": 1000,
            "초코붕어빵": 1200,
            "슈크림붕어빵": 1500
        }
        self.sales = {
            "팥붕어빵": 0,
            "초코붕어빵": 0,
            "슈크림붕어빵": 0
        }
    
    def stock(self):
        for bread, quantity in self.check_stock.items():
            print(f'{bread}: {quantity}개')
    
    def order(self):
        while True:
            B_type = input('주문하실 붕어빵을 입력해주세요(종료시:"종료"): ')
            if B_type == "종료":
                print("주문을 종료합니다.")
                break
            if B_type in self.check_stock:
                count = int(input('주문하실 개수를 입력해주세요: '))
                
                if count <= self.check_stock[B_type]:
                    self.check_stock[B_type] -= count
                    self.sales[B_type] += count
                    print(f'{B_type} {count}개 주문하셨습니다.')
                else:
                    print('주문할 수 있는 재고가 부족합니다.')
            else:
                print('존재하지 않는 붕어빵 종류입니다.')
        
        print(f'현재 재고: {self.check_stock}')

    def manager(self):
        while True:
            choose = input('1번 "재고관리" or 2번 "종료"')

            if choose == "1":
                B_type = input("추가할 붕어빵을 입력하세요")
                if B_type in self.check_stock:
                    add = int(input("추가할 개수를 입력해주세요"))
                    self.check_stock[B_type] += add
                    print(f'{B_type}{add}개를 추가했습니다.')
                else:
                    print("해당 붕어빵은 메뉴에 없습니다.")
            elif choose == "2":
                print("관리자 모드를 종료합니다.")
                break
    def total_sales(self):
        total = 0
        for bread, count in self.sales.items():
            total += count * self.prices[bread]
        return total
    
def main():
    Bshop = BungeoppangShop()
    while True:
        확인 = input('1번 "재고확인", 2번 "주문", 3번 "재고관리" 4번 "총 매출 확인" 5번 "종료"')
        if 확인 == "1":
            Bshop.stock()
        elif 확인 == "2":
            Bshop.order()
        elif 확인 == "3":
            Bshop.manager()
        elif 확인 == "4":
            print(f'총 매출: {Bshop.total_sales()}원')
        elif 확인 == "5":
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 입력입니다. 다시 시도해주세요.")

if __name__ == "__main__":
    main()
    


```
