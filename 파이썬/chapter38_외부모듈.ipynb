```python
from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen("https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnld=108")

soup = BeautifulSoup(target, "html.parser") # parser로 통해서 분석해서 컴퓨터가 이해할수 있는 언어로 알려줌

for location in soup.select("location"): # 로케이션에 해당하는거 싹 다 가져와서 for문에 집어 넣음
    print("도서:", location.select_one("city").string) # 다 가져온것 중 하나만 선택
    print("날씨:", location.select_one("wf").string)
    print("최저기온:", location.select_one("tmn").string)
    print("최고기온:", location.select_one("tmx").string)
    print()
```
