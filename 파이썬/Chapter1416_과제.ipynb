[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
주문내용 = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}

붕어빵 = input('주문할 붕어빵을 적어주세요: ')
개수 = int(input('주문하실 개수를 적어주세요: '))

# 재고 체크 및 주문 처리
if 붕어빵 in 주문내용:
    if 개수 <= 주문내용[붕어빵]:
        주문내용[붕어빵] -= 개수  # 주문한 만큼 재고 감소
        print(f'{붕어빵} {개수}개를 주문하셨습니다.')
        print(f'현재 재고: {주문내용}')
    else:
        print('주문할 수 있는 재고가 부족합니다.')
else:
    print('해당 붕어빵은 메뉴에 없습니다.')
```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
주문내용 = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}

while True:
    붕어빵 = input('주문할 붕어빵을 입력하세요 (종료: "종료"): ')
    
    if 붕어빵 == "종료":
        print("주문을 종료합니다.")
        break
    
    if 붕어빵 in 주문내용:
        개수 = int(input('주문하실 개수를 입력하세요: '))
        
        if 개수 <= 주문내용[붕어빵]:
            주문내용[붕어빵] -= 개수  # 주문한 만큼 재고 감소
            print(f'{붕어빵} {개수}개를 주문하셨습니다.')
        else:
            print('주문할 수 있는 재고가 부족합니다.')
    else:
        print('해당 붕어빵은 메뉴에 없습니다.')
    
    print(f'현재 재고: {주문내용}')
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
주문내용 = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}

def 관리자모드():
    while True:
        print("\n===== 관리자 모드 =====")
        print("1. 재고 추가")
        print("2. 종료")
        선택 = input("메뉴를 선택하세요 (1 또는 2): ")
        
        if 선택 == "1":
            붕어빵 = input('추가할 붕어빵을 입력하세요: ')
            if 붕어빵 in 주문내용:
                추가개수 = int(input('추가할 개수를 입력하세요: '))
                주문내용[붕어빵] += 추가개수
                print(f'{붕어빵} {추가개수}개를 추가하였습니다.')
            else:
                print('해당 붕어빵은 메뉴에 없습니다.')
        elif 선택 == "2":
            print("관리자 모드를 종료합니다.")
            break
        else:
            print("잘못된 입력입니다. 다시 입력해주세요.")

def 주문처리():
    while True:
        붕어빵 = input('주문할 붕어빵을 입력하세요 (종료: "종료"): ')
        
        if 붕어빵 == "종료":
            print("주문을 종료합니다.")
            break
        
        if 붕어빵 in 주문내용:
            개수 = int(input('주문하실 개수를 입력하세요: '))
            
            if 개수 <= 주문내용[붕어빵]:
                주문내용[붕어빵] -= 개수  # 주문한 만큼 재고 감소
                print(f'{붕어빵} {개수}개를 주문하셨습니다.')
            else:
                print('주문할 수 있는 재고가 부족합니다.')
        else:
            print('해당 붕어빵은 메뉴에 없습니다.')
        
        print(f'현재 재고: {주문내용}')

# 메인 코드
while True:
    print("\n===== 붕어빵 주문 시스템 =====")
    print("1. 주문하기")
    print("2. 관리자 모드")
    print("3. 종료")
    선택 = input("메뉴를 선택하세요 (1, 2, 3): ")
    
    if 선택 == "1":
        주문처리()
    elif 선택 == "2":
        관리자모드()
    elif 선택 == "3":
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# 초기 설정
가격 = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

판매개수 = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

총매출 = 0

def 주문처리():
    global 총매출
    
    while True:
        붕어빵 = input('주문할 붕어빵을 입력하세요 (종료: "종료"): ')
        
        if 붕어빵 == "종료":
            print("주문을 종료합니다.")
            break
        
        if 붕어빵 in 가격:
            개수 = int(input('주문하실 개수를 입력하세요: '))
            
            if 개수 <= 주문내용[붕어빵]:
                주문내용[붕어빵] -= 개수  # 주문한 만큼 재고 감소
                판매개수[붕어빵] += 개수  # 판매 개수 업데이트
                매출 = 가격[붕어빵] * 개수
                총매출 += 매출
                print(f'{붕어빵} {개수}개를 주문하셨습니다. 매출: {매출}원')
            else:
                print('주문할 수 있는 재고가 부족합니다.')
        else:
            print('해당 붕어빵은 메뉴에 없습니다.')
        
        print(f'현재 재고: {주문내용}')
        print(f'현재 판매 개수: {판매개수}')
        print(f'총 매출: {총매출}원')

# 메인 코드
while True:
    print("\n===== 붕어빵 주문 시스템 =====")
    print("1. 주문하기")
    print("2. 관리자 모드")
    print("3. 종료")
    선택 = input("메뉴를 선택하세요 (1, 2, 3): ")
    
    if 선택 == "1":
        주문처리()
    elif 선택 == "2":
        관리자모드()
    elif 선택 == "3":
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")

        
```
